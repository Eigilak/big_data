## 6
Select the movies that were released between 2010 and 2013, have a rating above 8.5, and have revenue above the 75th percentile.


```python
condition = (im['Year'] >= 2010) & (im['Year'] <= 2013) & (im['Rating'] > 8.5) & (im["Revenue (Millions)"] > im["Revenue (Millions)"].quantile(0.25))
im[condition]
```

## 9
Produce a dataframe consisting of the director, and the average rating for all the films with that director.


```python
im2 = im[['Director','Rating']].groupby('Director').mean().reset_index()
```

```python
im2.head()
```

## 10

```python
sorted = im[['Director','Rating']].sort_values('Rating', ascending = False).groupby('Director')
```
